## Exercise 4
Show me with examples of different arrays, the Broadcasting Rule of thumb: "Arrays can be broadcast if their dimensions match or if one of the arrays is 1".

```python
# Array broadcasting is a sensible way of doing elementwise operations on arrays of different (but compatible) shapes, such as:
arr2d = np.array([[0,1,2],[3,4,5]])
arr1d = np.array([1,2,3])
sum_arr = arr2d + arr1d
print(f'Given that the first and second array have a compatible shape of {arr2d.shape} and {arr1d.shape}, we can do an elementwise operation of this arrays, obtaining: \n {sum_arr}')
print('It works with plus, minus, times, exponentiation, min/max, and many more elementwise operations.')
```

## Exercise 5
Use Indexing to extract the values of a column and a row from the array. And add up their values.

```python
print(f'Given \n {arr_5d}')
print(f'the values from the 1st column are:{arr_5d[:,0]}, which add to {arr_5d[:,0].sum()}.')
print(f'the values from the 1st row are:{arr_5d[0,:]}, which add to {arr_5d[0,:].sum()}.')
```

## Exercise 6
Mask the above matrix, perform a vectorized Boolean calculation, taking each element and checking if it is evenly divided by four.

This returns a mask array in the same way as the elementary results of the calculation.

```python

mask = arr_5d % 4 == 0
print(f'Mask to check if the elements are divisible by 4: \n {mask}')
```

## Exercise 7
Then use this mask to index the original number array. This causes the array to lose its original shape, reducing it to one dimension, but you still get the data you are looking for.

```python
print(f'We can obtain all the values that satisfy the masks: {arr_5d[mask]}')
```
